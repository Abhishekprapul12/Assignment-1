```python
# Task 1: Discount Rules

order_input = input("Enter order amount: ")

# Check for numeric input
if order_input.isdigit():
    order_amount = int(order_input)
else:
    print("Error: Please enter a valid numeric amount.")
    exit()

# Apply discount rules
if order_amount >= 2000:
    discount_rate = 0.15
elif order_amount >= 1500:
    discount_rate = 0.10
elif order_amount >= 1000:
    discount_rate = 0.07
else:
    discount_rate = 0.0

discount = order_amount * discount_rate
subtotal = order_amount - discount

# Extra: 5% tax
tax = subtotal * 0.05
final_total = subtotal + tax

print("Original Amount:", order_amount)
print("Discount:", discount)
print("Subtotal after discount:", subtotal)
print("Tax (5%):", tax)
print("Final Total:", final_total)
```

```python
# Task 2: Process Multiple Orders

orders = [1200, 2500, 800, 1750, 3000]

total_revenue = 0
discount_count = 0

for order in orders:

    if order >= 2000:
        discount_rate = 0.15
    elif order >= 1500:
        discount_rate = 0.10
    elif order >= 1000:
        discount_rate = 0.07
    else:
        discount_rate = 0.0

    discount = order * discount_rate
    final_amount = order - discount

    print(order, "->", discount_rate * 100, "% ->", final_amount)

    total_revenue += final_amount

    if discount > 0:
        discount_count += 1

print("Total Revenue After Discounts:", total_revenue)
print("Orders With Discount:", discount_count)
```

```python
# Task 3: User Menu

orders = []

while True:

    print("\nMenu:")
    print("1 - Add Order")
    print("2 - Show Orders with Discounts")
    print("q - Quit")

    choice = input("Enter choice: ")

    if choice == "1":
        amount_input = input("Enter order amount: ")

        if amount_input.isdigit():
            orders.append(int(amount_input))
            print("Order added.")
        else:
            print("Invalid amount.")
            continue

    elif choice == "2":

        if len(orders) == 0:
            print("No orders available.")
            continue

        total = 0

        for order in orders:

            if order >= 2000:
                rate = 0.15
            elif order >= 1500:
                rate = 0.10
            elif order >= 1000:
                rate = 0.07
            else:
                rate = 0.0

            discount = order * rate
            final_amount = order - discount

            print(order, "->", final_amount)

            total += final_amount

        print("Total Revenue:", total)

    elif choice == "q":
        print("Exiting program...")
        break

    else:
        print("Invalid option. Try again.")
        continue
```

```python
# Task 4: Loop Control

daily = [200, 150, 0, 400, 50, -1, 300]

total_sales = 0

for sale in daily:

    if sale == -1:
        print("Corrupted data found. Stopping...")
        break

    if sale == 0:
        print("No sales today. Skipping...")
        continue

    total_sales += sale
    print("Running Total:", total_sales)

print("Final Total Sales:", total_sales)
```
