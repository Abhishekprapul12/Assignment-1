Product Pricing (Dictionaries)

```python
#Task 3: Product Pricing(Dictionaries)
#1. Create price dictionary with atleast 8 products
price_dict={'Jeanspant':2500, 'Bracelat':150000, 'Shoes':6000, 'Cellphone':30000, 'Teddybear':800, 'Medicine':500, 'Books':300, 'Magzines':250
}

#2.Add a new product
price_dict['PhotoFrames']=1200

#Update price of existing product
price_dict["Jeweleries"]=170000

# Remove product by name(handle if not exists)
product_to_remove="Magzines"

if product_to_remove in price_dict:
    del price_dict[product_to_remove]
    print(product_to_remove,"removed Successfully.")
else:
    print(product_to_remove,"does not exist.")

#5. Print average price
total=sum(price_dict.values())
count=len(price_dict)
average=total/count

print("Average price:",average)

#Extra:Product with maximum and minimum price
max_product=max(price_dict,key=price_dict.get)
min_product=min(price_dict,key=price_dict.get)

print("Highest priced product:",max_product,price_dict[max_product])
print("Lowest priced product:",min_product,price_dict[min_product])
```

Combined Operations

```python
#4. Combined operations
products=['Jeanspant','Bracelat','Shoes','Cellphone','Teddybear','Medicine']

categories=['Dress', 'Electronics', 'Electronics',"Dress",'Electronics',"Dress"]

price_dict={'Jeanspant':2500, 'Bracelat':150000, 'Shoes':6000, 'Cellphone':30000, 'Teddybear':800, 'Medicine':500, 'Books':300, 'Magzines':250
}

#1. create catalog list of tuples
catalog=[]

for i in range(len(products)):
    product=products[i]
    price=price_dict[product]
    category=categories[i]
    catalog.append((product,price,category))

print('catalog:',catalog)

# Create category_to_products dictinory
category_to_product={}

for product,price,category in catalog:
    if category not in category_to_product:
        category_to_product[category]=[]
    category_to_product[category].append(product)

print("Category mapping:",category_to_product)

#3. find category with maximum number of products

max_category=max(category_to_product,key=lambda x:len(category_to_product[x]))
print("Category with most product:",max_category)
print("Products in that category:",category_to_product[max_category])

```
